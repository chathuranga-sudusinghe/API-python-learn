````python
import pandas as pd
import requests

URL = "https://en.wikipedia.org/wiki/List_of_largest_banks"

headers = {
    "User-Agent": "Mozilla/5.0"
}

# Step 1: Get HTML content (browser-like request)
response = requests.get(URL, headers=headers)

# Step 2: Read ALL tables from the page
tables = pd.read_html(response.text)

# Step 3: Select the 2nd table (index starts from 0)
df_country = tables[1]

# Step 4: Display the table
df_country.head()
````
